```python
df_t = pd.read_csv('data/Transcript.csv', skiprows=range(1, 180000)).dropna(axis=0, subset=['End', 'PersonNumber', 'IdSubject'])
#sub_df = df_t.loc[df_t['End'].apply(lambda x: x.startswith('2015-11-30'))]
sub_df = df_t
```

```python
sub_df.shape
```

```python
def combine(l, n):
    # gets all permutated n-uples out of the list l
    return list(itertools.combinations(l, n))

def get_id_code(number):
    # needs person.csv as ppl
    return ppl.loc[ppl.PersonNumber == number, 'PersonIdCode'].tolist()[0]

def get_people_ids(df):
    # returns a list of unique ids, of the people speaking in a df
    return [int(get_id_code(x)) for x in df['PersonNumber'].unique().tolist()]

def update_adj(adj, pair):
    # updates adjacency matrix
    # creates weighted adj matrix
    try:
        adj.loc[pair[0], str(pair[1])] += 1
        adj.loc[pair[1], str(pair[0])] += 1
        print('adj updated')
    except:
        print('adj: not an active pair')
        
ppl = pd.read_csv('data/Person.csv')
test_df = sub_df
adj = pd.read_csv('data/adj.csv').set_index('PersonNumber')
subjects = test_df['IdSubject'].unique().tolist()
```

```python
subjects
```

```python
len(subjects)
```

```python
start_time = time.time()
for subj in subjects:
    one = test_df.loc[test_df.IdSubject == subj]
    people = get_people_ids(one)
    pairs = combine(people, 2)
    if len(pairs) > 0:
        for pair in pairs:
            print(pair)
            update_adj(adj, pair)
end_time = time.time()
print('time elapsed: '+str(end_time-start_time))
```
