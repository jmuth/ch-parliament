```python
def count_sessions(row):
    sesh = row['FirstSession']
    print(sesh)
    if sesh > 0:
        for i, j in enumerate(sessions):
            if int(j) == sesh:
                print('bitch')
                n = len(sessions)-i
                return int(n)
    else:
        pass
```

```python
ppl.fillna(0)
```

```python
ppl['NbrSessions'] = ppl['NbrSessions'].apply(lambda x: int(x) if (x > 0) else 0)
```

```python
def int_per_session(row):
    if row['NbrSessions'] > 0:
        return row['TotalInterventions']/row['NbrSessions']
    else:
        return 0
```

```python
ppl['InterventionPerSession'] = ppl.apply(int_per_session, axis=1)
```
