```python
import pandas as pd
import json
```

```python
ppl = pd.read_csv('new_data/people_jonas.csv').set_index('PersonIdCode')
```

```python
ppl.head(2)
```

```python
dico = {'names': []}
for row in ppl.iterrows():
    idcode = row[0]
    first = row[1].FirstName
    last = row[1].LastName
    name = first+' '+last
    dico['names'].append(name)
    dico[name] = idcode.item()
```

```python
dico
```

```python
with open('new_data/names.json', 'w') as outfile:
    json.dump(dico, outfile)
```
