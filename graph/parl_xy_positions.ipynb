```python
import pandas as pd
from bs4 import BeautifulSoup as BSoup
```

```python
people = pd.read_csv('new_data/people_jonas.csv').set_index('PersonIdCode')
```

```python
people['ParlPosX'] = 0
```

```python
people['ParlPosY'] = 0
```

```python
people.loc['3084']
```

```python
with open('new_data/parliament.html', 'r') as html:
    r = html.read()
soup = BSoup(r, 'lxml')
```

```python
circles = soup.find_all('circle')
```

```python
for circle in circles:
    people.loc[int(circle['id']), 'ParlPosX'] = circle['cx']
    people.loc[int(circle['id']), 'ParlPosY'] = circle['cy']
```
