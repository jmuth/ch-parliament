```python
import pandas as pd
import json
```

# Computing top friends

```python
adj = pd.read_csv('new_data/new_adja.csv').set_index('PersonNumber')
```

```python
dico = {}
for i in adj.index:
    row = adj.loc[i].sort_values(ascending=False)
    friends = []
    incr = 0
    for j, k in row.iteritems():
        if incr > 4:
            break
        sub_dico = {'friend': j, 'number': k.item()}
        friends.append(sub_dico)
        incr += 1
    dico[i.item()] = friends
```

```python
dico
```

```python
with open('new_data/friends.json', 'w') as outfile:
    json.dump(dico, outfile)
```
