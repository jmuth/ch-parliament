```python
import pandas as pd
import json
```

# Computing top friends from adjacency matrix

```python
adj = pd.read_csv('good_data/adj_intervention.csv').set_index('PersonNumber')
```

```python
dico = {}
for i in adj.index:
    row = adj.loc[i].sort_values(ascending=False)
    friends = []
    incr = 0
    for j, k in row.iteritems():
        if k.item() > 0:
            sub_dico = {'friend': j, 'number': k.item()}
            friends.append(sub_dico)
    dico[i.item()] = friends
```

```python
with open('good_data/friends.json', 'w') as outfile:
    json.dump(dico, outfile)
```
