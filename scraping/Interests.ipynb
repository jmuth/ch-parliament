```python
import json
def prepare_d3(in_1, out_):
    interest = pd.read_csv(in_)
    interest.head()
    cols = np.array(interest.columns)[:-1]
    interest_json = {}
    for i in range(len(interest)):
        id_ = int(interest.PersonIdCode[i])
        array = []
        for i in themes.keys():
            val_json = {}
            val_json["int"] = int(interest[interest.PersonIdCode == id_][i])
            val_json["theme"] = themes[i]["english"]
            val_json["idx"] = themes[i]["val"]
            array.append(val_json)
        interest_json[str(id_)] = array
    with open(out_, "w") as outfile:
        json.dump(interest_json, outfile)   
        
def prepare_hist(in_1, in_2, out_):
    interest1 = pd.read_csv(in_1)
    interest2 = pd.read_csv(in_2)
    cols = np.array(interest.columns)[:-1]
    interest_json = {}
    for i in range(len(interest1)):
        id_ = int(interest1.PersonIdCode[i])
        array = []
        for i in themes.keys():
            val_json = {}
            val_json["auth"] = int(interest1[interest1.PersonIdCode == id_][i])
            val_json["cos"] = int(interest2[interest2.PersonIdCode == id_][i])
            val_json["theme"] = themes[i]["english"]
            val_json["idx"] = themes[i]["val"]
            array.append(val_json)
        interest_json[str(id_)] = array
    with open(out_, "w") as outfile:
        json.dump(interest_json, outfile)
```

```python
df_business = pd.read_csv('data/Business.csv')
df_role = pd.read_csv('data/BusinessRole.csv')
temp = pd.read_csv('data/Tags.csv')
active = pd.read_csv('data/ActivePeople.csv')
```

```python
# get tag for each business
df_business_tags = df_business[['ID', 'Tags']].set_index('ID')
df_business_tags = df_business_tags.dropna(axis=0)
```

```python
# create a DataFrame business->tag filled with zero
topics = pd.DataFrame(index=df_business_tags.index)
for i in temp.ID:
    topics[i] = 0

# split the tag "4|8|12" in a list(4, 8, 12)
df_business_tags['Tags'] = df_business_tags['Tags'].apply(lambda x: x.split('|'))

# fill the cell of topics table
def fill_cell(business_number, tags_array):
    for tag in tags_array:
        previous_val = topics.get_value(int(business_number), int(tag))
        topics.set_value(int(business_number), int(tag), int(previous_val + 1))

for i, tags in zip(df_business_tags.index, df_business_tags.Tags):
    fill_cell(i, tags)

# ensure topics has type integer
topics = topics.astype(int)
```
